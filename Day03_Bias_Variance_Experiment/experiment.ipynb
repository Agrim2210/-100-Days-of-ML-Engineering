```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df=pd.read_csv("dataset.csv")
X=df[["x"]].values
y=df["y"].values
```

```python
#Degree=1(Underfitting,High bias,low Variance)

Poly=PolynomialFeatures(degree=1)
x_poly=Poly.fit_transform(X)
model=LinearRegression()
model.fit(x_poly,y)
y_pred=model.predict(x_poly)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y,color="blue",label="data")

plt.plot(X,y_pred,color="red",label="degree 1 model")

plt.title("Underfitting Example (Polynomial Degree = 1)")
plt.legend()
plt.savefig("Underfitting_Visualisation.png")

plt.show()
```

```python
#Degree=3 (Accurate fit ,ie-low bias,low variance)
Poly=PolynomialFeatures(degree=3)
x_poly=Poly.fit_transform(X)
model.fit(x_poly,y)
y_pred_3=model.predict(x_poly)
```

```python

plt.scatter(X,y,color="blue",label="data")

plt.plot(X,y_pred_3,color="red",label="degree 3 model")
plt.title("Accurate fit (degree=3)")

plt.legend()
plt.savefig("Accurate_fit_Visualisation.png")

plt.show()
```

```python
#Degree=15(Overfit,ie high variance,low bias)
Poly=PolynomialFeatures(degree=15)
x_poly=Poly.fit_transform(X)
model=LinearRegression()
model.fit(x_poly,y)
y_pred_15=model.predict(x_poly)
```

```python

plt.scatter(X,y,color="blue",label="data")

plt.plot(X,y_pred_15,color="red",label="degree 15 model")

plt.title("Overfitting Example (Polynomial Degree = 15)")

plt.legend()
plt.savefig("Overfitting_Visualisation.png")

plt.show()
```
