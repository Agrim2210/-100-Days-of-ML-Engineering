```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
df = pd.read_csv("dataset.csv")
y = df["price"]
```

```python
X_base = df[["size_sqft","rooms"]]
X_train,X_test,y_train,y_test=train_test_split(X_base,y,test_size=0.20,random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred_base = model.predict(X_test)

print("R2 Score (Base Features):", r2_score(y_test, y_pred_base))


```

```python
df["size_rooms"] = df["size_sqft"] * df["rooms"]

X_interaction = df[["size_sqft","rooms","size_rooms"]]
X_train,X_test,y_train,y_test=train_test_split(X_interaction,y,test_size=0.20,random_state=42)

model.fit(X_train, y_train)

y_pred_interaction = model.predict(X_test)

print("R2 Score (Interaction Features):", r2_score(y_test, y_pred_interaction))

```
