```python
#Load Dataset
import pandas as pd 
df=pd.read_csv("dataset.csv")
X=df.drop("Price_lakh",axis=1).values
y=df["Price_lakh"].values
X=(X-X.mean(axis=0))/X.std(axis=0)
print(X,y)
```

```python
#Train Model 
from linear_regression import LinearRegression
model=LinearRegression(lr=0.01,epochs=10000)
model.fit(X,y)
predictions=model.predict(X)
print(predictions)
```

```python
import matplotlib.pyplot as plt
plt.plot(model.loss_history)
plt.title("Loss convergence")
plt.xlabel("Iterations")
plt.ylabel("MSE Loss")
plt.savefig("loss_curve.png")
plt.show()
```

```python
# Comparing with sklearn -Linear Regression 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
sk=LinearRegression()
sk.fit(X,y)
sk_predict=sk.predict(X)
print("Scratch MSE:", mean_squared_error(y, predictions))
print("Sklearn MSE:", mean_squared_error(y, sk_predict))
```
