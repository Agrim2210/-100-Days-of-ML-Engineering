```python
import pandas as pd 
df=pd.read_csv("dataset.csv")
X=df.drop("Bought_Laptop",axis=1).values
y=df["Bought_Laptop"].values
X=(X-X.mean(axis=0))/(X.std(axis=0))
X
```

```python
import numpy as np
from Logistic_Regression import LogisticRegression
model=LogisticRegression(lr=0.01,epochs=10000)
model.fit(X,y)
pred=model.predict(X)

```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y,pred))
```

```python
from sklearn.linear_model import LogisticRegression
sk=LogisticRegression()
sk.fit(X,y)
sk_pred=sk.predict(X)
print(" SKlearn Accuracy:",accuracy_score(y,sk_pred))
```

```python
import matplotlib.pyplot as plt
plt.plot(model.loss_history)
plt.title("Log Loss Convergence")
plt.xlabel("Iterations")
plt.ylabel("Binary Cross Entropy")
plt.savefig("Log_loss_convergence.png")
plt.show()
```
