```python
import pandas as pd
df=pd.read_csv("dataset.csv")
X=df.drop("price",axis=1).values
y=df["price"].values
```

```python
# Raw model ie Without Scaling
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
model_raw=LinearRegression()
model_raw.fit(X,y)
```

```python
#With Standardization
Scale=StandardScaler()
X_tranf=Scale.fit_transform(X)
model_trans=LinearRegression()
model_trans.fit(X_tranf,y)
```

```python
print(model_raw.coef_)
print(model_trans.coef_)

```

```python
# Applied Gradient descent to visualize convergence
from gradient_descent import gradient_descent
loss_raw=gradient_descent(X,y)
loss_scaled=gradient_descent(X_tranf,y)
```

```python
#Plotting loss curve 
import matplotlib.pyplot as plt
plt.plot(loss_raw,label="Without Scaling")
plt.plot(loss_scaled,label="with Scaling")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Gradient Descent Convergence")
plt.legend()
plt.savefig("Scaling_convergence.png")
plt.show()
```
