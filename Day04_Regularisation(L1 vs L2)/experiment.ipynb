```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge,Lasso
df=pd.read_csv("dataset.csv")
X=df.drop("y",axis=1).values
y=df["y"]
```

```python
#Linear Regression(No Regularisation)
lr=LinearRegression()
lr.fit(X,y)
lr.coef_
```

```python
#L2(Ridge Regression)
ridge=Ridge(alpha=2)
ridge.fit(X,y)
ridge.coef_
```

```python
#L1(lasso regression)
lasso=Lasso(alpha=2)
lasso.fit(X,y)
lasso.coef_
```

```python
#Coefficient Comparison Visualisation
coefs = np.array([
lr.coef_,
ridge.coef_,
lasso.coef_
])

labels = ["x1","x2","x3"]

x = np.arange(len(labels))

plt.bar(x-0.2,coefs[0],width=0.2,label="Linear")
plt.bar(x,coefs[1],width=0.2,label="Ridge")
plt.bar(x+0.2,coefs[2],width=0.2,label="Lasso")

plt.xticks(x,labels)
plt.title("Coefficient Comparison")
plt.legend()
plt.savefig("Coefficient_Comparison_Visualisation.png")

plt.show()
```
