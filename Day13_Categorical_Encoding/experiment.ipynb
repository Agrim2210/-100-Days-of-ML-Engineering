```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#Label Encoding
df = pd.read_csv("dataset.csv")
y = df["price"]
le=LabelEncoder()
df["city_label"]=le.fit_transform(df["city"])
X_label=df[["city_label","rooms"]]
```

```python
model = LinearRegression()
model.fit(X_label, y)

y_pred_label = model.predict(X_label)

print("R2 Score (Label Encoding):", r2_score(y, y_pred_label))
```

```python
#One Hot Encoding
df_ohe = pd.get_dummies(df, columns=["city"], drop_first=True)
X_ohe = df_ohe.drop(columns=["price","city_label"])

model.fit(X_ohe, y)

y_pred_ohe = model.predict(X_ohe)

print("R2 Score (One Hot Encoding):", r2_score(y, y_pred_ohe))
```
